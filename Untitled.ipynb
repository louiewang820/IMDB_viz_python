```python
import pandas as pd
import dash
from dash import dcc
from dash import html
from dash.dependencies import Input, Output, State

# Load data
movies = pd.read_csv("data/imdb_top_1000.csv", dtype={'Runtime': str})

# Data wrangling function
def wrangled_data(df, minRevenue, year, runtimes, genre, star):
    df = df.copy()
    df['Runtime'] = df['Runtime'].str.extract(r'(\d+)')
    df['Gross'] = pd.to_numeric(df['Gross'].str.replace(',', '')) / 1000000
    df = df[df['Gross'] >= minRevenue]
    df = df[(df['Released_Year'] >= year[0]) & (df['Released_Year'] <= year[1])]
    df = df[(df['Runtime'] >= runtimes[0]) & (df['Runtime'] <= runtimes[1])]
    df = df[df['Genre'].str.contains('|'.join(genre))]
    df = df[(df['Star1'].isin(star)) | (df['Star2'].isin(star)) | (df['Star3'].isin(star)) | (df['Star4'].isin(star))]
    return df
```
